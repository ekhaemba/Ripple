```
countryIDs = {
    0:"World",
    4:"Afghanistan",
    528:"Netherlands"
    }
def chunks(l, n):
    """Yield successive n-sized chunks from l."""
    for i in range(0, len(l), n):
        yield l[i:i + n]
def getData(r=[0],px="HS",ps=2016,p=[0],rg="all",cc=[],max=500,type="C",freq="A",fmt="csv", head="M"):
  for country in r:
    url = "https://comtrade.un.org/api/get?r={}&px={}&ps={}&p={}&rg={}&cc={}&max={}&fmt={}&type={}&freq={}&head={}"
    partnerLists = list(chunks(p,5))
    HSLists = list(chunks(cc,5))
    for partnerList in partnerLists:
      for HSList in HSLists:
        codedURL = url.format(country,px,ps,"%2C".join(map(str,partnerList)),rg,"%2C".join(map(str,HSList)),max,fmt,type,freq,head)
        print(codedURL)
        resp = requests.get(codedURL)
        text = resp.text
        time.sleep(1)
        if fmt.lower() == "csv":
          yield StringIO(text)
        else:
          yield resp.json()   
def doQuery (conn):
  cur = conn.cursor()
  cur.execute("SELECT * FROM Blockchain.cocoapuffs WHERE cmdCode=1701")
  for data in cur.fetchall():
    print(data)
```

```
def storeIntoFrame(**kargs):
  df, new_frame = None, None
  #r=[276],p=[528,251,56,384,616,566],cc=[1701,1801,1804],rg="all",fmt="json",head="M"
  for data in getData(**kargs):
    new_frame = pd.DataFrame(data['dataset'])
    df = pd.concat([df, new_frame])
  return df    
```

```
total_records = my_frame[my_frame['ptCode'] == 0]
world_records = my_frame[my_frame['ptCode'] != 0]
```

```
#1 for imports, 2 for exports
total_cocoa_import = total_records[total_records['rgCode'] == 1]['TradeQuantity']
total_cocoa_export = total_records[total_records['rgCode'] == 2]['TradeQuantity']
cocoa_imports = world_records[world_records['rgCode'] == 1]
cocoa_exports = world_records[world_records['rgCode'] == 2]
```

```
cocoa_imports.loc[:,'Percentage'] = (cocoa_imports['TradeQuantity']/int(total_cocoa_import))*100
cocoa_exports.loc[:,'Percentage'] = (cocoa_exports['TradeQuantity']/int(total_cocoa_export))*100
```

```
cocoa_imports.sort_values('Percentage',ascending=False)[:10]
```
